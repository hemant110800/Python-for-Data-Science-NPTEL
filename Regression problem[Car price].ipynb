# Working range of data

```python
cars=cars[(cars.yearOfRegistration<=2018)&(cars.yearOfRegistration>=1950)&(cars.price>=100)&(cars.price<=150000)&(cars.powerPS>=10)&(cars.powerPS<=500)]
cars
```
